```python
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
def scat_to_azi(scat_ang,sza,vza=40):
    '''
    compute relative azimuth for given scattering angle, solar zenith and viewing zenith angles.
    :param scat_ang: scattering angle in deg.
    :param sza: solar zenith angle in deg.
    :param vza: viewing zenith angle in deg.
    :return: relative azimuth in deg.
    '''
    #cos(ang) = -np.cos(szar) * np.cos(vzar) - np.sin(szar) * np.sin(vzar) * np.cos(azir)
    vzar=np.radians(vza)
    scat_angr=np.radians(scat_ang)
    szar=np.radians(sza)
    azir =np.arccos( (np.cos(scat_angr) +np.cos(szar) * np.cos(vzar) )/(np.sin(szar) * np.sin(vzar)))
    return np.degrees(azir)
```

```python
rc = {'axes.labelsize': 20,'font.size': 18,'xtick.labelsize':18,'ytick.labelsize':18}
plt.rcParams.update(rc)
fig, axs = plt.subplots( figsize=(7, 5))
szas= np.linspace(0.1,88,300)
plt.plot(szas,scat_to_azi(130,szas,vza=40),color='black',ls=':',label='$130^\circ$')
plt.plot(szas,scat_to_azi(135,szas,vza=40),color='black',label='$135^\circ$')

plt.plot(szas,scat_to_azi(140,szas,vza=40),color='black',ls='--',label='$140^\circ$')
plt.xlabel(r'$Sun\ zenith\ angle\ (deg.)$')
plt.ylabel(r'$Relative\ azimuth\ (deg.)$')
plt.minorticks_on()
plt.legend(title='$Scat.\ ang.$',fontsize=14)
plt.tight_layout()
```
